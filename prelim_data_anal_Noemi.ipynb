```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./test/SongCSV.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
fig = plt.figure()
ax = df.plot.scatter(x = 'Year', y = "Tempo")
ax.set_xlim(1910, 2015)
ax.set_title("Year vs Tempo")
ax.set_ylabel("Tempo (bpm)")
# plt.savefig('year_vs_tempo.png')
plt.show()
```

```python
df_new = df[(df['SongHotttnesss'] > 0) & (df['ArtistHotttnesss'] > 0) & np.isfinite(df['ArtistHotttnesss'])]
```

```python
df_new.shape
```

```python
df_new.plot.scatter('ArtistHotttnesss', 'SongHotttnesss')
```

```python
song_grouped = df_new.groupby('ArtistHotttnesss').SongHotttnesss.mean()
art_grouped = df_new.groupby('ArtistHotttnesss').ArtistHotttnesss.mean()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(art_grouped, song_grouped, 'o')
ax.set_title("Artist Hotttnesss vs Song Hotttnesss")
ax.set_xlabel("Artist Hotttnesss")
ax.set_ylabel("Song Hotttnesss")
#plt.savefig('artisthotttnesss_vs_songhotttnesss.png')
plt.show()
```
